```python
import numpy as np
import pandas as pd
data=pd.read_csv('kmeans.csv')
x=np.array(data)[:,:-1]
y=np.array(data)[:,-1]
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3,random_state=0).fit(x,y)
z=kmeans.predict([['0.906','0.606']])
print(z)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'kmeans.csv' contains your data
data = pd.read_csv('kmeans.csv')
x = np.array(data.iloc[:, :-1])  # Use iloc to select all rows and all columns except the last one
y = np.array(data.iloc[:, -1])   # Use iloc to select all rows and only the last column

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=0).fit(x)
new_data_point = np.array([[0.906, 0.606]])

# Add a new column to your DataFrame with the predicted clusters
data['Cluster'] = kmeans.labels_

# Plot the data points with different colors for each cluster
plt.scatter(x[:, 0], x[:, 1], c=data['Cluster'], cmap='viridis', edgecolor='k')
plt.scatter(new_data_point[0, 0], new_data_point[0, 1], marker='X', s=200, c='red', label='New Data Point')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='o', s=300, c='blue', label='Cluster Centers')

plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
